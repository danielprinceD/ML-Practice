```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
ds = pd.read_csv("spam_ham_dataset.csv")
```

```python

ds.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ds['text'] = ds['text'].astype('category')
ds['text_cod'] = ds['text'].cat.codes
```

```python
encoder = OneHotEncoder()
```

```python
encode_data = pd.DataFrame(encoder.fit_transform(ds[['text_cod']]).toarray())
```

```python
ds.join(encode_data)
```

```python
log = LogisticRegression()
ds.iloc[0]
```

```python
data = pd.DataFrame(encode_data)
data = data.join(ds['label_num'])
```

```python
x = np.array(data.iloc[0::])
y = np.array(data['label_num'])
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.25)
```

```python
log.fit(x_train,y_train)
```

```python
y_pred = log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```
