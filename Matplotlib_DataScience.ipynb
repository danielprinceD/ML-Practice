```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(0,50,0.2)
b1 = np.sin(a)
b2 = a**2+2*a
```

```python
plt.plot(a,b2)
```

```python
plt.plot(a,b1)
```

```python
a1 = plt.subplot(121)
a2 = plt.subplot(122)
a1.plot(a,b1)
a2.plot(a,b2)
plt.tight_layout()
```

```python
b3 = np.cos(a)

plt.plot(a,b1,'r',label="sine")
plt.plot(a,b3,'g',label="cosine")
plt.legend(loc="upper right")
plt.grid(True)
plt.tight_layout()
```

```python
python = (90,60,20)
java =  (10,70,40)
html = (90,50,45)

person = ['one','two','three']

index = np.arange(3)

plt.bar(index,python,width=0.1,label="Python")
plt.bar(index+0.2,java,width=0.1,label="Java")
plt.bar(index+0.4,java,width=0.1,label="HTML")

plt.xlabel("Person")
plt.ylabel("Percentage")
plt.xticks(index+0.2,person)
plt.legend(loc="upper left")
plt.ylim(0,150)
```

PIECHART


```python
state = ['TamilNadu','Andhra','Kerala','Other']
stud  = [90,40,30,20]
explode = [0.01,0.01,0.01,0.01]
a = plt.pie(stud,labels=state,autopct="%0.2f%%",explode=explode)
```

```python
anime = ["One Piece","AOT","Naruto"]
rate = [90,85,75]
both = zip(anime,rate)
both = sorted(list(both))
anime,rate = zip(*both)
plt.pie(rate,labels=anime,autopct="%0.2f%%",explode=[0.01,0.01,0.01],counterclock = False,startangle=90,colors=['green','orange','red'])
plt.title("Top Animes")
```

```python
a = 172 + 10* np.random.randn(10000)
plt.hist(a,100,facecolor='b',density=True)
plt.xlabel("Student Height(in cm)")
plt.ylabel("Percentage(in %)")
plt.tight_layout()
plt.grid(True)
plt.text(140,0.031,"Avg = 172, S.D = 10")
plt.show()
```

#3D-Plotting


```python
from mpl_toolkits import mplot3d
z = np.linspace(0,100,1000)
x = np.sin(z)
y = np.cos(z)
axes = plt.axes(projection = "3d")
axes.plot3D(x,y,z)
plt.show()
```

#3D-Surface Plot

```python
def z_function(x,y):
  return np.sin(np.sqrt(x**2+y**2))
x = np.linspace(-15,15,50)
y = np.linspace(-15,15,50)
X,Y = np.meshgrid(x,y)
Z = z_function(X,Y)
axes = plt.axes(projection="3d")
axes.plot_surface(X,Y,Z)
plt.show()
```
